```python
# Importing all necessary libraries
import pandas as pd
import seaborn as sns
```

```python
# Importing the dataset
# The dataset is assumed to be in the same directory as this script
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
# Time to clean up the dataset
# Say we only want USA Races, 50K OR 50mI, 2020
```

```python
#Show 50k
df[df["Event distance/length"] == '50km']
```

```python
# Combine 50k and 50mI with isin
df[df["Event distance/length"].isin(['50km', '50mi'])]
```

```python
#Now we only want to see the ones in 2020
df[df["Event distance/length"].isin(['50km', '50mi']) & (df["Year of event"] == 2022)]
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'HKG']
```

```python
# Now we want to combine all the filters together
df[df["Event distance/length"].isin(['50km', '50mi']) & (df["Year of event"] == 2022) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'HKG') ]
```

```python
df2 = df[df["Event distance/length"].isin(['50km', '50mi']) & (df["Year of event"] == 2022) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'HKG') ]
```

```python
# Want to remove the 'HKG' from the event name
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
# Cleaning up the athlete data
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
# remove the h from Athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(10)
```

```python
#remove h from athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
# Time to clean up null values
df2.isnull().sum()
```

```python
#Dropping columns that are irrelevant: Athlete Club, Athlete Country, Athlete year of birth, Athlete Age Category
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
# Now we want to clean up null values
df2[df2['athlete_age'].isnull()]
```

```python
df2 = df2.dropna(subset=['athlete_age'])
df2.isnull().sum()
df2.shape
```

```python
#Check for duplicates values
df2[df2.duplicated() == True] 
```

```python
# Reset index
df2.reset_index(drop=True)
```

```python
# Fixing the types
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
# Renaming Columns
df2 = df2.rename(columns={
    'Year of event': 'year',
    'Event dates': 'date',
    'Event name': 'eventName', 
    'Event distance/length': 'distance', 
    'Event number of finishers' : 'num_of_finishers',
    'Athlete performance': 'performance', 
    'Athlete gender' : 'gender',
    'Athlete ID' : 'athleteID',
    'Athlete average speed': 'avg_speed', 
    'athlete_age': 'age'
})
```

```python
df2.head(10)
```

```python
# reorder the columsn
df3 = df2[['date', 'year', 'eventName', 'distance', 'num_of_finishers', 'athleteID','performance', 'avg_speed', 'age', 'gender']]
```

```python
# Drop the year column since all the races are from 2022
df3 = df3.drop(['year'], axis=1)
```
